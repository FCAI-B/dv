```python
import pandas as pd

df0 = pd.read_csv('canada-mig-dataset.csv')

df0.head()
```

```python
df1 = df0.iloc[:, 9:43]
df1.head()
```

```python
df1.loc["Total"] = df1.sum(axis=0)
df1.tail()
```

```python
df2 = df1.tail(1)
df2.head()
```

```python
df3 = df2.transpose()
df3.head()
```

```python
df3.reset_index(inplace=True)
df3.head()
```

```python
df3.columns = ['Year', 'Total']
df3.head()
```

```python
df3.plot(kind='scatter', y='Total', x='Year', figsize=(16, 6))
```
